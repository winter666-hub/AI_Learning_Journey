# Cost Function

```python
import numpy as np
X_train = np.array([
    [2104,5,1,45],
    [1416,3,2,40],
    [852,2,1,35]
], dtype=float)

y_train = np.array([460,232,178], dtype=float)
```

# Gradient

```python
def compute_gradient(X, y, w, b):
    m,n = X.shape

    dj_dw = np.zeros(n)
    dj_db = 0

    for i in range(m):
        error = np.dot(X[i], w) + b -y[i]

        for j in range(n):
            dj_dw[j] += error * X[i,j]
        
        dj_db += error

    dj_dw = dj_dw/m
    dj_db = dj_db/m

    return dj_dw, dj_db
```

# Gradient Descent

```python
def gradient_descent(X, y, w, b, alpha, iterations):
    for i in range(iterations):
        dj_dw, dj_db = compute_gradient(X, y, w, b)

        w = w - alpha*dj_dw
        b = b - alpha*dj_db

    return w, b
```

```python
import numpy as np

def compute_cost(X, y, w, b):
    m = X.shape[0]
    preds = X @ w + b
    return (1/(2*m)) * np.sum((preds - y)**2)

def gradient_descent_verbose(X, y, w, b, alpha, iterations, print_every=1):
    for it in range(1, iterations + 1):
        dj_dw, dj_db = compute_gradient(X, y, w, b)

        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        if it % print_every == 0:
            cost = compute_cost(X, y, w, b)
            print(f"iter {it:>3} | cost={cost:>10.6f} | b={b:>10.6f} | w={w}")

    return w, b

# 초기값
w_init = np.zeros(X_train.shape[1])
b_init = 0.0

# 발산 가능성이 있으므로 작게 시작
w_final, b_final = gradient_descent_verbose(
    X_train, y_train,
    w=w_init, b=b_init,
    # 숫자가 크므로 학습률을 0.0000001로 설정
    alpha=1e-7, iterations=20, print_every=1
)

print("\nFinal parameters:")
print("w_final =", w_final)
print("b_final =", b_final)

preds = X_train @ w_final + b_final
print("\nPredictions vs y:")
for i in range(len(y_train)):
    print(f"example {i}: pred={preds[i]:.3f}, y={y_train[i]:.3f}")
```
