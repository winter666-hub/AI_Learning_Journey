# 02. Cost Function - linear Regression

## Definition

J(w,b) = 1/(2m) Σ (wx + b - y)^2

비용 합수는 모델의 예측값과 실제값의 차이를 제곱하여 평균낸 값이다.  
모델이 데이터를 얼마나 잘 맞추는지를 수치로 나타낸다.

# 비용 함수 구현

```python
import numpy as np
import matplotlib.pyplot as plt

x_train = np.array([1.0, 2.0, 3.0])
y_train = np.array([1.0, 2.0, 3.0])

def compute_cost(x, y, w, b):
    """
    Compute cost function for linear regression.
    
    J(w,b) = (1/2m) Σ (f(x) - y)^2
    """
    m = x.shape[0]
    cost = 0

    for i in range(m):
        f_wb = w * x[i] + b
        cost += (f_wb - y[i]) ** 2

    total_cost = cost / (2 * m)
    return total_cost
```

# 여러 w 값 실험

```python
w_values = [0, 0.5, 1]
b = 0

for w in w_values:
    cost = compute_cost(x_train, y_train, w, b)
    print(f"w = {w}, Cost={cost}")
```

# Cost 시각화

```python
# w를 바꿔가면서 비용이 어떻게 변하는지 계산
w_range = np.linspace(-2, 3, 100)
cost_values = []

for w in w_range:
    cost_values.append(compute_cost(x_train, y_train, w, 0))

plt.plot(w_range, cost_values)
plt.xlabel("w")
plt.ylabel("J(w)")
plt.title("Cost Function vs w")
plt.show()
```

## Observations

- Cost is always >= 0  제곱이기 때문에 항상 0 이상이다.

- The graph is convex (U-shaped) 비용 함수는 볼록 함수이며 최소값이 하나 존재한다.

- The minimum occurs at w = 1 for this dataset  현재 데이터에서는 w = 1 일 때 최소이다.
