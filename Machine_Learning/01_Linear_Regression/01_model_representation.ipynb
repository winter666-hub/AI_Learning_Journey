# 01. Model Representation - Linear Regression

## Goal
Understand how a linear model represents data

선형 회귀 모델은 다음과 같은 형태를 가진다.

f(x) = wx + b

- w : weight (기울기)
- b : bias (절편)

# 데이터 정의

```python
import numpy as np
import matplotlib.pyplot as plt

# Training data (훈련 데이터)
x_train = np.array([1.0, 2.0, 3.0])
y_train = np.array([1.0, 2.0, 3.0])

# m = number of training examples
m = x_train.shape[0]
print("Number of training examples:", m)
```

# 모델 구현

```python
def compute_model_output(x, w, b):
    """
    Compute predictions using linear model.
    f(x) = wx + b
    
    x : input feature
    w : weight (기울기)
    b : bias (절편)
    """
    m = x.shape[0]
    f_wb = np.zeros(m)

    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb
```

# 실험 및 시각화

```python
w = 1
b = 0

predictions = compute_model_output(x_train, w, b)

plt.plot(x_train, predictions, label ="Prediction")
plt.scatter(x_train, y_train, color = 'red', label = "Actual")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

# 관찰 정리

## Observations (관찰 결과)

- w controls the slope of the line. w가 커질수록 직선이 가팔라진다.

- b shifts the line vertically. b는 직선을 위아래로 이동시킨다.

- when w = 1 and b = 0, the model perfectly fits the data. 이 데이터에서는 완벽히 맞는다.
