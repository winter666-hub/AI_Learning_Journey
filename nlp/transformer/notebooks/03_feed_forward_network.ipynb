```python
import torch
from torch import nn
```

```python
class FeedForwardNetwork(nn.Module):
    def __init__(self, d_model=512, d_ff=2048):
        super().__init__()

        # 첫 번째 Linear Layer
        # d_model(512) -> d_ff(2048)
        # 더 풍부한 특징을 학습하기 위해 차원을 확장
        self.linear1 = nn.Linear(d_model, d_ff)

        # 비선형성을 추가하는 활성화 함수
        self.relu = nn.ReLU()

        # 두 번째 Linear Layer
        # d_ff(2048) -> d_model(512)
        # 원래 차원으로 다시 압축
        self.linear2 = nn.Linear(d_ff, d_model)

    # FFN은 각 토큰을 독립적으로 변환하며, 토큰 간 정보는 Attention에서 처리한다.
    def forward(self, x):
        x = self.linear1(x)
        x = self.relu(x)
        x = self.linear2(x)

        return x
```

```python
ffn = FeedForwardNetwork()
x = torch.randn(32, 10, 512)
output = ffn(x)
print(output.shape)
```
