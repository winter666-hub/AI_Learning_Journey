```python
import torch
from torch import nn
```

```python
class AddNorm(nn.Module):
    def __init__(self, d_model=512):
        super().__init__()

        # nn.LayerNorm(d_model): 512개의 벡터들을  
        self.layer_norm= nn.LayerNorm(d_model)

    def forward(self, x, sublayer_output):
        x = x + sublayer_output
        x = self.layer_norm(x)

        return x
```
