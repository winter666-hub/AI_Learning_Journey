```python
import torch

seq_len = 4
mask = torch.tril(torch.ones(seq_len, seq_len))
print(mask)
```

```python
scores = torch.tensor([
    [2., 5., 8., 1.],
    [1., 4., 7., 3.],
    [3., 2., 6., 4.],
    [1., 5., 2., 8.]
])

scores = scores.masked_fill(mask==0, float('-inf'))

print(scores)
```

```python
import torch.nn.functional as F

attention = F.softmax(scores, dim=-1)

print(attention)
```
