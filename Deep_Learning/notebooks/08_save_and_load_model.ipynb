# Save and Load the Model

## 학습내용
- torch.save()를 이용한 모델 저장
- torch.load()를 이용한 모델 불러오기
- state_dict()를 이용한 모델 파라미터 저장
- 저장된 모델로 추론(inference) 수행
- model.eval()를 이용한 평가 모델 전환

```python
import torch
import torchvision.models as models
```

## Save and loading model weights

```python
model = models.vgg16(weights='IMAGENET1K_V1')
# state_dict() 모델의 weight, bias를 딕셔너리 형태로 저장한 것
torch.save(model.state_dict(), '../models/model_weights.pth')
```

```python
# VGG16 모델 구조 생성
model = models.vgg16()
# 저장된 모델 파라미터(weight, bias) 불러오기
# weights_only=True 모델 객체나 임의의 Python 객체는 복원하지 말고
# Tensor(weight, bias)만 읽어라 보안 상 문제 방지
model.load_state_dict(torch.load('../models/model_weights.pth', weights_only=True))
# 추론을 위한 평가 모드 시작
model.eval()
```

```python
# 모델 구조, 파라미터 전체 저장
torch.save(model, '../models/model.pth')

# 모델 구조, 파라미터 전체 불러오기
model = torch.load("../models/model.pth", weights_only=False)

# 추론을 위한 평과 모드 전환
model.eval()
```

torch.save(model)은 모델 전체를 저장하는 방식이다.
실무에서는 torch.save(model.state_dict())를 더 많이 사용한다.

## 정리
- torch.save()를 이용해 모델을 저장할 수 있다.
- torch.load()를 이용해 저장된 모델을 불러올 수 있다.
- state_dict()는 모델의 학습 파라미터(weight, bias)를 저장한다.
- model.load_state_dict()로 저장된 파라미터를 모델에 적용한다.
- model.eval()은 추론 전에 호출해야 하먀 Dropout과 BatchNorm을 평가 모드로 전환한다.
- torch.save(model.state_dict()) 방식이 PyTorch 권장 방식이다.
- torch.save(model)은 모델 구조와 파라미터를 함께 저장하지만 pickle에 의존하므로 실무에서는 잘 사용하지 않는다.

## 핵심
- 저장: torch.save()
- 불러오기: torch.load()
- 추론 전: model.eval()
- 실무: state_dict 저장/불러오기 사용
