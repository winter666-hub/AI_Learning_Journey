# PyTorch Transforms

## 학습 내용
- transform을 이용한 입력 데이터 변환
- target_transform을 이용한 label 변환
- One-Hot Encoding
- Compose를 이용한 transform 조합 
- Lambda를 이용한 사용자 정의 transform

```python
import torch
import torch.nn.functional as F
from torchvision import datasets
from torchvision.transforms import v2

ds = datasets.FashionMNIST(
    root="../data",
    train=True,
    download=True,
    transform=v2.Compose([v2.ToImage(), v2.ToDtype(torch.float32, scale=True)]),
    # FashionMNIST의 label(정수)을 Pytorch Tensor로 변환
    # One-Hot Encoding 수행 (FashionMNIST의 클래스는 10개)
    # one_hot()의 결과는 정수형(int64)이므로 float32로 변환
    target_transform=v2.Lambda(
        lambda y: F.one_hot(torch.tensor(y), num_classes=10).float()
    ),
)
```

## 정리

- transform은 입력 데이터를 변환할 때 사용한다.
- target_transform은 label을 변환할 때 사용한다
- Compose로 여러 transform을 순서대로 적용할 수 있다.
